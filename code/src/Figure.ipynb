```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure()
plt.bar(np.arange(1,6), np.array([3190, 2146, 4539, 11422, 48703]))
plt.xlabel('rating')
plt.ylabel('number')
plt.title('Rating count')
plt.savefig('./RatingCount.png',format='png')
plt.show()
```
